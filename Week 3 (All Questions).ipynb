## Libraries Done & Now Getting the site details for scrapping 

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url,'lxml')
```

```python
table_post = soup.find('table')
fields = table_post.find_all('td')

postcode = []
borough = []
neighbourhood = []

for i in range(0, len(fields), 3):
    postcode.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighbourhood.append(fields[i+2].text.strip())
        
data = pd.DataFrame(data=[postcode, borough, neighbourhood]).transpose()
data.columns = ['Postcode', 'Borough', 'Neighbourhood']
```

```python
data.head(10)
```

## Removing N/A values

```python
data['Borough'].replace('Not assigned', np.nan, inplace=True)
data.dropna(subset=['Borough'], inplace=True)
data.head()
```

```python
#reset index
updated = data.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
updated.columns = ['Postcode', 'Borough', 'Neighbourhood']
updated
```

```python
updated['Neighbourhood'].replace('Not assigned', "Queen's Park", inplace=True)
updated
```

# Question 2:- Getting the longitude and Latitude

---

```python
geo = pd.read_csv('http://cocl.us/Geospatial_data')
geo.columns = ['Postcode', 'Latitude', 'Longitude']
```

## Merging the tables Updated and Geo Tables

```python
position = pd.merge(updated, geo, on=['Postcode'], how='inner')
df_tor = position[['Borough', 'Neighbourhood', 'Postcode', 'Latitude', 'Longitude']].copy()
df_tor
```

# Question 3:- Generating the Toronto Map and making markdown

---

```python
warning="ignore"
address = 'Toronto, Canada'
geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
#generating map from folium
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)
final=zip(df_tor['Latitude'], df_tor['Longitude'], df_tor['Borough'], df_tor['Neighbourhood'])
```

```python
#adding details

for lat, lng, borough, neighborhood in final:
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=4,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
